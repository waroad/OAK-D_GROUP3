```python
import wave
import sys
from vosk import Model, KaldiRecognizer
import json
import speech_recognition as sr
import difflib
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import time
```

```python
# Offline Speech to text by vosk. (https://github.com/alphacep/vosk-api)
# Vosk is an offline open source speech recognition toolkit. 
# It is still being updated (there was update in last month too.) and has more than 5k stars, so 
# we chose this api to have offline Speech to text.

with open('texts.json', 'r', encoding='utf-8') as openfile:
    json_object = json.load(openfile)
    
off_start = time.time()
for i in range(1,51):
    wf = wave.open('ex_{}.wav'.format(i), "rb")

    if wf.getnchannels() != 1 or wf.getsampwidth() != 2 or wf.getcomptype() != "NONE":
        print("Audio file must be WAV format mono PCM.")
        sys.exit(1)

    model = Model(lang="en-us")
    rec = KaldiRecognizer(model, wf.getframerate())
    rec.SetWords(True)
    rec.SetPartialWords(True)

    while True:
        data = wf.readframes(4000)
        if len(data) == 0:
            break
        if rec.AcceptWaveform(data):
            rec.Result()
        else:
            rec.PartialResult()

    json_result=json.loads(rec.FinalResult())
    json_object["offline-"+str(i)]=json_result["text"]
    # print(json_result)
    print(json_result["text"])
off_end = time.time()
json_object = json.dumps(json_object, indent=4)
with open("texts.json", "w") as outfile:
    outfile.write(json_object)
```

```python
# Online Speech to text by Uberi's speech recognizer. (https://github.com/Uberi/speech_recognition)
# It has over 6k stars and also it is being updated currently (last updated 2 months ago) so we chose this.
# We specifically used 'Google Speech Recognition' from the api.

with open('texts.json', 'r', encoding='utf-8') as openfile:
    json_object = json.load(openfile)

r = sr.Recognizer()
on_start = time.time()

for i in range(1,51):
    with sr.AudioFile('ex_{}.wav'.format(i)) as source:
        audio_text = r.listen(source)

        # recoginize_() method will throw a request error if the API is unreachable, hence using exception handling
        try:
            text=r.recognize_google(audio_text)
        except:
            print("Online didn't work.")
    json_object["online-" + str(i)] =text
    print(i)
on_end = time.time()

json_object = json.dumps(json_object, indent=4)
with open("texts.json", "w") as outfile:
    outfile.write(json_object)
```

```python
# We used 3 different methods to check how well it detected.


with open('texts.json', 'r', encoding='utf-8') as openfile:
    json_object = json.load(openfile)

    
# Cosine Similarity 
cosine={"online":[], "offline":[]}
for i in range(1,51):
    sentences = (json_object["answer-{}".format(i)], json_object["online-{}".format(i)],
                 json_object["offline-{}".format(i)])

    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
    cos_similar = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    # print("online: ",cos_similar[0][0])
    cosine["online"].append(cos_similar[0][0])
    cos_similar = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[2:3])
    # print("offline: ", cos_similar[0][0])
    cosine["offline"].append(cos_similar[0][0])

# Jaccard Similarity
jaccard={"online":[], "offline":[]}
for i in range(1,51):
    answer_string = json_object["answer-{}".format(i)]
    input_string1 = json_object["online-{}".format(i)]
    input_string2 = json_object["offline-{}".format(i)]

    intersection_cardinality = len(set.intersection(*[set(answer_string), set(input_string1)]))
    union_cardinality = len(set.union(*[set(answer_string), set(input_string1)]))
    similar = intersection_cardinality / float(union_cardinality)
    # print("online: ",similar)
    jaccard["online"].append(similar)

    intersection_cardinality = len(set.intersection(*[set(answer_string), set(input_string2)]))
    union_cardinality = len(set.union(*[set(answer_string), set(input_string2)]))
    similar = intersection_cardinality / float(union_cardinality)
    # print("offline: ",similar)
    jaccard["offline"].append(similar)

# SequenceMatcher
sequenceMatcher={"online":[], "offline":[]}
for i in range(1,51):
    answer_bytes = bytes(json_object["answer-{}".format(i)], 'utf-8')
    input_bytes1 = bytes(json_object["online-{}".format(i)], 'utf-8')
    input_bytes2 = bytes(json_object["offline-{}".format(i)], 'utf-8')
    answer_bytes_list = list(answer_bytes)
    input_bytes1_list = list(input_bytes1)
    input_bytes2_list = list(input_bytes2)

    sm = difflib.SequenceMatcher(None, answer_bytes_list, input_bytes1_list)
    similar1 = sm.ratio()
    sm = difflib.SequenceMatcher(None, answer_bytes_list, input_bytes2_list)
    similar2 = sm.ratio()

    # print("online:",similar1)
    sequenceMatcher["online"].append(similar1)
    # print("offline:",similar2)
    sequenceMatcher["offline"].append(similar2)

df = pd.DataFrame(list(zip(cosine["offline"], cosine["online"],jaccard["offline"], jaccard["online"]
                          ,sequenceMatcher["offline"], sequenceMatcher["online"])),
               columns =['cos_off', 'cos_on', 'jac_off', 'jac_on', 'seq_off', 'seq_on'])

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

colors = ['b','r','b','r','b','r']

fig,ax = plt.subplots(figsize=(5,5)) 
ax,props = df.plot.box(patch_artist=True, return_type='both', ax=ax)
for patch,color in zip(props['boxes'],colors):
    patch.set_facecolor(color)

ax.set_title("The Difference in Accuracy between Online and Offline Environments") # set the title of the boxplot    
ax.set_ylabel("Accuracy Difference") # set the y-axis label    
ax.set_xlabel("Environment") # set the x-axis label of the boxplot
    
plt.show()
```

 The graphs of cos, jac, and seq were analyzed when executed both online and offline, with the x-axis representing the type and the y-axis representing the accuracy. The results indicate that all three functions demonstrated higher accuracy levels when executed online compared to offline.
